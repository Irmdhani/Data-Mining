# Dataset Load

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('datasets-jadi.csv')
```

```python
df.head()
```

# Proses Datasets

```python
df.rename(index=str, columns={
    'RIW HIPERTENSI_Ya' : 'HIPERTENSI',
    'RIW PE_Ada' : 'PE',
    'OBESITAS_Ya' : 'OBESITAS',
    'RIW DM_Ya' : 'DM',
    'RIW HIPERTENSI/PE DALAM KELUARGA_Ada' : 'HIPER/PE DLM KEL',
}, inplace=True)
```

```python
X = df.drop(['NO','NAMA','RIW HIPERTENSI_Tidak',
             'RIW PE_Tidak','OBESITAS_Tidak','RIW DM_Tidak',
             'RIW HIPERTENSI/PE DALAM KELUARGA_Tidak','SOSEK RENDAH_<UMR'], axis=1)
```

```python
print(X)
```

# menentukan jumlah cluster

```python
clusters = []
for i in range(1,11):
    km = KMeans(n_clusters=i).fit(X)
    clusters.append(km.inertia_)

fig, ax = plt.subplots(figsize=(12,8))
sns.lineplot(x=list(range(1,11)), y=clusters, ax=ax)
ax.set_title('mencari elbow')
ax.set_xlabel('clusters')
ax.set_ylabel('inertia')
```

# Clustering

```python
n_clust = 4
kmean = KMeans(n_clusters=n_clust).fit(X)
X['Clusters'] = kmean.labels_
```

```python
plt.figure(figsize=(10,8))

sns.scatterplot(x=X['USIA'], y=X['PARITAS'], hue=X['Clusters'], 
                size=X['Clusters'], palette=sns.color_palette('hls', n_clust))

                
```
